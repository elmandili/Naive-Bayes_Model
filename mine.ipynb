```python
import pandas as pd
```

```python
london = pd.read_json('london.json', lines = True)
paris = pd.read_json('paris.json', lines=True)
new_york = pd.read_json('new_york.json', lines = True)
random_tweets = pd.read_json('random_tweets.json', lines = True)


```

```python
london_text = london['text'].tolist()
paris_text = paris['text'].tolist()
new_york_text = new_york['text'].tolist()

tweets = london_text + paris_text + new_york_text
labels = [0]*len(london_text) + [1]*len(paris_text) + [2]*len(new_york_text)

```

```python
from sklearn.model_selection import train_test_split

train_data, test_data,train_labels, test_labels = train_test_split(tweets, labels, test_size=0.2, random_state=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

counter = CountVectorizer()
counter.fit(train_data)

train_counts = counter.transform(train_data)
test_counts = counter.transform(test_data)
```

```python
from sklearn.naive_bayes import MultinomialNB

classifier = MultinomialNB()
classifier.fit(train_counts, train_labels)
score = classifier.score(test_counts, test_labels)
print(score)
```
